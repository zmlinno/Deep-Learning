## 激活函数

```python
max(1,0)
#max 是返回括号里面的最大值
```

```python
# 没有激活函数的话，神经网络不会拐弯，不会拐弯的话处理不了非线性问题
# 激活函数， relu，sigmoid
# 用的最多的激活函数 relu：
def relu(x):
    return max(x,0)
#如果 x<0 返回0，如果大于0就会返回自己
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,[relu (i) for i in x])
```

```python
# simiod 一般作为分类任务最后一层即输出层的激活函数
# 一般不做隐层的激活函数，因为容易梯度消失或梯度爆炸
```

```python
# tanh 双曲正切函数
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
plt.plot(x,[tanh (i) for i in x])
```

```python
# softplus
def softplus(x):
    return np.log(np.exp(x)+1)
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,[softplus (i) for i in x])
```
